```python
import matplotlib.pyplot as plt
import numpy as np

# 设置绘图风格为 fivethirtyeight
plt.style.use('fivethirtyeight')

# 使用 numpy 生成示例数据：x 是一个从 0 到 10 的 100 个均匀分布的点
x = np.linspace(0, 10, 100)
# y1 和 y2 分别是这些点的正弦和余弦值
y1 = np.sin(x)
y2 = np.cos(x)
print(y1, y2)
# 创建图形和坐标轴对象，使用 figsize=(10, 6) 设置图形大小为 10x6 英寸,1英寸严格等于2.54厘米
# 如果你使用 plt.subplots(nrows=2, ncols=3)，这将创建一个 2x3 的子图网格，总共包含 6 个子图
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))

# 绘制两条线
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

# 设置图表标题和坐标轴标签，使用中文以提高可读性。
ax.set_title("Sine and cosine functions")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")

# 添加图例,显示每条线的标签,右上角
ax.legend()

# 添加网格线
ax.grid(True)

# 使用 plt.tight_layout() 自动调整子图参数，以给定的填充和间距为每个轴留出空间
plt.tight_layout()

# 最后，使用 plt.show() 显示图表
plt.show()
```
